```python
import pandas as pd
import matplotlib.pyplot as plt

# 예시 데이터프레임 생성
data = {'A': [1, 2, 3, 4, 5],
        'B': [5, 4, 3, 2, 1]}
df = pd.DataFrame(data)

# 산점도 그리기
plt.scatter(df['A'], df['B'])

# 그래프에 제목과 레이블 추가
plt.title('Scatter Plot')
plt.xlabel('A')
plt.ylabel('B')

# 그래프 표시
plt.show()
```
