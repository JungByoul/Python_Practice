```python
import pandas as pd

df = pd.read_excel('2024-2_과목별_20250120.xlsx')

df.head(5)


```

```python
import pandas as pd
from openpyxl import Workbook
from openpyxl.styles import Alignment, Font, Border, Side

df.columns = [col.replace(' ', '_') for col in df.columns]

required_columns = ['학과_및_전공', '단과_및_석/박', '사번', '성명', '직종', '영역', '학수번호', '교과목명', '학점', '수강신청\n인원', '강의평가\n적용인원', '강의평가']
data = df[required_columns]
data.rename(columns={'성명':'이름', '수강신청\n인원' :'수강인원', '강의평가\n적용인원':'평가인원'}, inplace=True)

data.head(5)
```

```python
depart_major_avg = data.groupby('학과_및_전공')['강의평가'].mean().round(2).reset_index()
depart_major_avg.rename(columns={'강의평가':'강의평가_학과_평균'}, inplace=True)


college_avg = data.groupby('단과_및_석/박')['강의평가'].mean().round(2).reset_index()
college_avg.rename(columns={'강의평가':'강의평가_단과대학_평균'}, inplace=True)


n_data = pd.merge(data, depart_major_avg, on='학과_및_전공', how='left')

# 원본 데이터프레임에 단과별 평균 병합
n_data = pd.merge(n_data, college_avg, on='단과_및_석/박', how='left')


depart_major_avg.head(10)
college_avg.head(10)
```

```python
n_data.head(10)
n_data = n_data.query("직종 not in ['교수', '조교수', '부교수']") 

n_data.head(30)
```

```python
from openpyxl.styles import Alignment, Font, Border, Side, PatternFill
import os
#이제 각 샘플 데이터 제작

# dep_set=set()
# for dep in data['학과_및_전공']:
#     dep_set.add(dep)
    
# table_data = n_data.drop(columns=['학과_및_전공', '단과_및_석/박', '강의평가_학과_평균', '강의평가_단과대학_평균'])
# table_data.head(5)

cnt= 0
for dep in n_data['학과_및_전공'].unique():
    cnt+=1
    
    dep =dep ##학과 이름
    department_avg = n_data[n_data['학과_및_전공'] == dep]['강의평가_학과_평균'].values[0]

    college = n_data[n_data['학과_및_전공'] == dep]['단과_및_석/박'].values[0]
    college_avg = n_data[n_data['단과_및_석/박'] == college]['강의평가_단과대학_평균'].values[0]
    
    if dep == '특강/분과/공통선택':
        continue
    
    # print(f"학과:{dep} 학과평균: {depart_major_avg} 단과 평균 {college_avg}")

    #학과별 데이터 필터링
    filtered_data = n_data[n_data['학과_및_전공'] == dep]
    table_data = filtered_data.drop(columns=['학과_및_전공', '단과_및_석/박', '강의평가_학과_평균', '강의평가_단과대학_평균'])

    #오름차순진행
    table_data = table_data.sort_values(by='이름', )
    
    # print(table_data)
    
    # Create a new Excel file
    wb = Workbook()
    ws = wb.active
    ws.title = "샘플 데이터"

    # A1:J1 Title
    ws.merge_cells('A1:J1')
    ws['A1'] = "2024학년도 2학기 강의평가 결과"
    ws['A1'].alignment = Alignment(horizontal='center', vertical='center')
    ws['A1'].font = Font(name='Arial', bold=True, size=14)

    # A2: College Name
    ws['A2'] = college
    ws['A2'].font = Font(bold=True)

    # A3: Department Name
    ws['A3'] = dep
    ws['A3'].font = Font(bold=True)

    # A4:J4 Averages
    ws.merge_cells('A4:J4')
    ws['A4'] = (
        f"학부 평균: 4.59 / "
        f"{college} 평균: {college_avg} / "
        f"{dep} 평균: {department_avg}"
    )
    ws['A4'].font = Font(color="FF0000FF")

    # 이제 표 시작!!

    # Headers for the table
    headers = table_data.columns.tolist()
    ws.append([''] * 5)  # 5행 전체를 비움
    ws.append(headers)  # 5번 열을 비우고 6번 열에서 시작

    for col in ws.iter_cols(min_row=6, max_row=6, min_col=1, max_col=len(headers)):
        for cell in col:
            cell.font = Font(bold=True, color="FF000000")  # 검은색 폰트
            cell.alignment = Alignment(horizontal='center', vertical='center')

    # 색상 설정 (색조 22, 채도 15%, 밝기 99%)
    fill_red = PatternFill(start_color="FFF7E6", end_color="FFF7E6", fill_type="solid")
    fill_gray = PatternFill(start_color="D9D9D9", end_color="D9D9D9", fill_type="solid")


    # Append data rows
    for row in table_data.itertuples(index=False):
        ws.append(row)

    # Add borders to all cells
    thin_border = Border(
        left=Side(style='thin'),
        right=Side(style='thin'),
        top=Side(style='thin'),
        bottom=Side(style='thin')
    )
    
        # 강의평가_평점 열의 인덱스 찾기
    eval_col_index = headers.index('강의평가') + 1

    for row in ws.iter_rows(min_row=7, max_row=ws.max_row, min_col=1, max_col=len(headers)):
        for cell in row:
            cell.border = thin_border
            if cell.column == eval_col_index:  # 강의평가 데이터 열에만 색상 적용
                cell.fill = fill_red
                cell.font = Font(bold=True)
                
    # for row in ws.iter_rows(min_row=7, max_row=ws.max_row, min_col=1, max_col=len(headers)+1):
    #     for cell in row:
    #         cell.border = thin_border
    #         if headers[cell.column-1] == '강의평가_평점':  # 강의평가 데이터 열에만 색상 적용
    #             cell.fill = fill_red
    #             cell.font = Font(bold=True)

    for col in ws.iter_cols(min_row=6, max_row= 6, min_col=1, max_col=len(headers)):
        for cell in col:
            cell.fill = fill_gray

    # Save the file
    output_path = f'2024-2_{college}_{dep}_.xlsx'
    

    # # 디렉토리가 존재하지 않으면 생성
    # if not os.path.exists(os.path.dirname(output_path)):
    #     os.makedirs(os.path.dirname(output_path), exist_ok=True)
    #->> 국제지역대학원 교학과  특강/분과/공통선택  학과는 에러나서 이거 제외하고 추출
    
    wb.save(output_path)
    

```
