```python
import pandas as pd

# CSV 파일 경로 지정
file_path = '/content/drive/MyDrive/python_project/1_목적지_데이터/youngduck_stay_time_2022.csv'

# CSV 파일 불러오기
df = pd.read_csv(file_path, encoding='cp949')

# 데이터프레임 확인
df
```

```python
#기본정보 확인하기
df.info()      #제목행 정리가되지 않아서, Dtype이 object로 처리됨

```

```python
#지역명 컬럼에서 영덕군만 필터링하기

column_name = '지역명'  # 기준으로 할 컬럼명 변수 생성
filtered_df = df[df[column_name] == '영덕군']           # 기준컬럼이 특정조건('내국인')을 만족하는 새로운 데이터프레임 생성

filtered_df



```

```python
#쓸모없는 데이터(열) 삭제해주기 (지역명 컬럼)

filtered_df.drop('지역명', axis=1, inplace=True)        #axis가 1이면 열, 0이면 행을 기준으로 함수를 호출함

filtered_df
```

```python
filtered_df.info()
```

```python
# 기준연월 컬럼을 인덱스로 바꿔주기

filtered_df.set_index(filtered_df.columns[0], inplace=True)
filtered_df
```

```python
#새로운 csv로 저장하기
filtered_df.to_csv('/content/drive/MyDrive/python_project/1_목적지_데이터/전처리완료/preprocessing_youngduck_stay_time_2022.csv', index=True, encoding='cp949')
#index = True 는 인덱스도 저장하는 옵션, False는 인덱스는 저장하지 않도록 하는 옵션
#데이터 프레임에 한글이 있을경우, 인코딩으로 해줘야함. utf-8이나, cp949 중에서 utf8는 오류가 발생함
```
